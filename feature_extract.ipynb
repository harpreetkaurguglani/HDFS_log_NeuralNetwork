```python
import featureprocessing as fp
import dataloader as dl
import preprocessing as prep
import pandas as pd
```

```python
data='HDFS.log_structured.csv'
labels='anomaly_label.csv'
```

```python
# load in the datasets
(x_train_pre, y_train),(x_test_pre, y_test)=dl.load_HDFS(data,label_file=labels,window='session',train_ratio=1,split_type='uniform',save_csv=True)
```

```python
# feature extractor
feature_extractor=prep.FeatureExtractor()
# feature extraction on training set
x_train=feature_extractor.fit_transform(x_train_pre,term_weighting=None,normalization=None)
```

```python
#feature extraction dataframe with labels
features_matrix=pd.DataFrame(x_train)
features_matrix['anomaly_label']=y_train
features_matrix.head()
```
